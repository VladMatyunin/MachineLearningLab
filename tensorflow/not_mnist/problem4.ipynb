```python
from merge_dataset import train_dataset, test_dataset, valid_dataset
from merge_dataset import train_labels, test_labels, valid_labels
import matplotlib.pyplot as plt
label_names = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
```

```python
def display_dataset_top20(dataset_name, dataset, labels):
    print("DATASET " + dataset_name)
    for i in range(0, 20):
        plt.imshow(dataset[i], interpolation='nearest')
        plt.show()
        print("Label: ", label_names[labels[i]])
    print("\n_______________________________________")


display_dataset_top20('test', test_dataset, test_labels)
display_dataset_top20('train', train_dataset, train_labels)
display_dataset_top20('valid', valid_dataset, valid_labels)
```
